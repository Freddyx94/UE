##LR - Scratch


```python
import numpy as np
```

Formula:

m = $\Sigma$(x-$\hat{x}$)(y-$\hat{y}$) / $\Sigma$(x-$\hat{x}$)Â²

b = $\hat{y}$ - m*$\hat{x}$

predict with $\hat{y}$ = m*x + b.

```python
def fit_linear_regression_1d(X, y): #calculate the slope
  x = np.asarray(X, dtype=float) #Convert to float
  y = np.asarray(y, dtype=float) #Convert to float

  x_mean = x.mean() #Mean
  y_mean = y.mean()

  #Calculate slope
  num =np.sum((x-x_mean) * (y-y_mean))
  den = np.sum((x-x_mean) ** 2)

#Intercept
  m = num / den
  b = y_mean - m * x_mean

  return m,b

def predict_linear_regression_1d(x,m,b):
  x = np.asarray(x, dtype=float)
  return m * x + b

  #Linear Regresion
```

```python
df_small = df.head(20) #First 20 Countries
x = df_small["LandArea"].values # 1D
y = df_small["Population"].values
```

```python
#X and y are 1D arrays here

m, b = fit_linear_regression_1d(x,y)

print(f"Slope (m): {m:.6f}")
print(f"Intercpt (b): {b:.2f}")
print()
print("Model formula:")
print(f"Predicted Population = {m:.6f} * LandArea + {b:.2f}")
```

```python
# Make prediction with our own function

y_pred = predict_linear_regression_1d(x, m, b)

results = df_small.copy()
results["Predicted_Population"] = y_pred.astype(int)

print("Actual vs Predicted Population (first 20 countries):")
display(results[["Country", "LandArea", "Population", "Predicted_Population"]])
```
